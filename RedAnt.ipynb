```python
n=int(input("Num of Relationship "))


'''family_tree is stored as a dictionary
Each person in the family_tree will have his own dictionary
{
  relationship1 : [name1,name2],
  relationship2 : [name1,name2,name3]
}'''

family=dict()
me = input("Enter your name ")

#Adding first member to the dict
family[me]=dict()

print("Enter the details in below format!")
print("Name1 Name2 Relationship (Make sure that Name1 is already in family tree)\n")

i=0
while i<n:
  text=input().split()

  if len(text) < 3: #to check if input has 3 words
    print("Invalid input!")
    continue
  elif text[0] not in family: #to check if Name1 is already present in dict
    print(text[0]+" doesn't exist in family tree")
    continue

  if text[2] not in family[text[0]]:  #to create a new list for the given relationship
    family[text[0]][text[2]]=[]
  if text[1] not in family:  #to add name2 to dict if it is not present
    family[text[1]]=dict()

  family[text[0]][text[2]].append(text[1])
  print(text[1]+" ("+text[2]+") of "+text[0]+" is added to family tree!")
  i=i+1

print(family)
```

```python
from graphviz import Digraph

#dot = Digraph(comment = 'Ancestry', graph_attr = {'splines':'ortho'})

dot = Digraph(comment = 'Family Tree', graph_attr = {'splines':'true','overlap':'scalexy','nodesep':'1'})
for person in family: #take each person in family dict
  for relation in family[person]: #take each relationship found in that particular person's dict
    for temp in family[person][relation]: #add a new edge to all the person found in the list
      dot.edge(person,temp,relation)

dot
```
